```python
import pandas as pd
import plotly.express as px
url='https://drive.google.com/file/d/1sHZCgx9Vs3LrLpe0kBzZK2kxZD3IWPdM/view'
url='https://drive.google.com/uc?id=' + url.split('/')[-2]
df = pd.read_excel(url)
display(df)
```

```python
copia = df.copy()
df2 = copia[['ANO','UFN','Município', 'ESPVIDA', 'IDHM']]

display(df2)
```

```python
df3 = df2.sort_values(by='IDHM', ascending=False).head(20)
display(df3)
```

```python
media_idhm = px.histogram(df3,x='Município', y='IDHM', histfunc="max")
media_idhm.update_yaxes(range=[0.5, 1])
media_idhm.show("png")
```

```python
#ESPVIDA = Esperança de vida ao nascer
espvida_idhm = px.scatter(copia, x='ESPVIDA', y='IDHM', color='IDHM', trendline='ols')
espvida_idhm.show("png")
```
